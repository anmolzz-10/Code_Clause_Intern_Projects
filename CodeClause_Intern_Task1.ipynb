```python
import pandas as pd
import seaborn as sns

from sklearn.datasets import load_iris
iris = load_iris()
iris_data = pd.DataFrame( iris.data, columns=iris.feature_names)
iris_data['target'] = iris.target
```

```python
sns.pairplot(iris_data, hue='target', markers=["o", "s", "D"])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for i, feature in enumerate(iris_data.columns[:-1]):
    plt.subplot(2, 2, i + 1)
    sns.histplot(data=iris_data, x=feature, kde=True, hue='target', multiple='stack')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(iris_data.corr(), annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```
